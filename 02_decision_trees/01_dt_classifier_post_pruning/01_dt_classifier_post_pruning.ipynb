```python
# Import dependicies

import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Load data

df = pd.read_csv("machine_health.csv")

df.head()
```

```python
# Define Input and Output Features

# X = df[['Temperature', 'Vibration', 'Current', 'Noise']]
X = df.drop('Machine_Status', axis=1) # -> Keep all columns except Machine_Stauts, axis = 1 move vertically
y = df['Machine_Status']
```

```python
# Train-Test Split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state= 42
)

print("Total Data Lenght:", len(X))
print("Train Data Lenght:", len(X_train))
print("Test Data Length:", len(X_test))

```

```python
# Train Model 

model = DecisionTreeClassifier(max_depth=4) #Post - Pruning
model.fit(X_train, y_train)
```

```python
# Predict on Test Data

y_pred = model.predict(X_test)
```

```python
# Evaluate Model

print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nClassification Report")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, y_pred))
```

```python
# Plot Decision Tree

plt.figure(figsize=(30,35))

plot_tree(
    model,
    feature_names=X.columns,
    class_names=model.classes_.astype(str),
    filled=True
)

plt.show()
```
